```python
data_1 = data.drop(['AddressLine1','PostalCode','PhoneNumber'],axis=1)
```

## VARIABLE HANDLING -DATA PREPROCESSING

 Selecting the top 5 values and then marking the other values as 'Other'

```python
top_5 = data_1['City'].value_counts()[0:5]
rem_1 = data_1['City'].value_counts().keys()[5:]
rem_1
```

For State Province Name - keeping the top 15 values

```python
rem_2 = data_1.StateProvinceName.value_counts().keys()[15:]
rem_2
```

```python
data_1 = data_1.replace({
    
    'StateProvinceName':rem_2
},'Other')
data_1.StateProvinceName.value_counts()## Rest values are replaced by other
```

## Encoding the Variables

#### Gender

```python
dummy_1 = pd.get_dummies(data_1['Gender'],drop_first=True)## Drop_first = True is written to avoid the concept of dummy variable trap
dummy_1
```

```python
new_data = pd.concat([dummy_1,data_1],axis=1)
new_data['Male'] = new_data['M']
new_data = new_data.drop(['M'],axis=1)
new_data
```

```python
new_data.MaritalStatus.unique()
```

## One Hot Encoding = Maritial Status

```python
dummy_2 = pd.get_dummies(new_data['MaritalStatus'],drop_first=True)
dummy_2
```

```python
new_data_1 = pd.concat([dummy_2,new_data],axis=1)
new_data_1['MaritalStatus-Single'] = new_data_1['S']
new_data_1 = new_data_1.drop(['S'],axis=1)
```

```python
#new_data_1 = new_data_1.drop(['Gender','MaritalStatus'],axis=1)
```

```python
new_data_1
```

```python
dummy_occ = pd.get_dummies(new_data_1['Occupation'],drop_first=True)
dumm_ed = pd.get_dummies(new_data_1['Education'],drop_first=True)
dummy_country = pd.get_dummies(new_data_1['CountryRegionName'],drop_first=True)

new_data_2 = pd.concat([dummy_occ,dumm_ed,dummy_country,new_data_1],axis=1)

new_data_2
```

```python
#new_data_2 = new_data_2.drop(['Education','Occupation','CountryRegionName'],axis=1)
new_data_2
```

```python
dummy_city = pd.get_dummies(new_data_2['City'],drop_first=True)
dummy_state = pd.get_dummies(new_data_2.StateProvinceName,drop_first=True)

new_data_3 = pd.concat([dummy_city,dummy_state,new_data_2],axis=1)
new_data_3 = new_data_3.drop(['City','StateProvinceName'],axis=1)
new_data_3
```

#### Now Finally dropping the first and last name columns

```python
datafile_final = new_data_3.drop(['FirstName','LastName'],axis=1)
```

```python
datafile_final
```

##### Now the variables -  'Number of cars owned' and 'NumberofchildrenAtHome' are cardinally encoded

## Standardization of the Dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

scaled = scaler.fit(datafile_final)
data_scaled = scaler.transform(datafile_final)

data_scaled
```

## Converting to a dataframe containing scaled values

```python
datafile_scaled = pd.DataFrame(data_scaled,columns=datafile_final.columns)
datafile_scaled
```
