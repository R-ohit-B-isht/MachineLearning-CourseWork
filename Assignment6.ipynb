```python
df.insert(0, "ones", 1, True)
n = df.shape[0]
alpha = 0.01
```

```python
df1 = df.copy()
df2 = df.copy()
df3 = df.copy()
```

```python
def ConvertToBinary(df,ind):
  df.is_copy = False
  for i in range(0,n):
    if df.iloc[i,5] == ind:
      df.iloc[i,5]  = 1 
    elif df.iloc[i,5] != ind:
      df.iloc[i,5]  = 0
  print(df)
  return df
```

```python
df1=ConvertToBinary(df1,0)
df2=ConvertToBinary(df2,1)
df3=ConvertToBinary(df3,2)
```

**REAL CODE STARTS HERE**

```python
beta_count = df.shape[1]-1
beta1 = np.zeros(beta_count)
beta2 = np.zeros(beta_count)
beta3 = np.zeros(beta_count)
```

```python
def findBetas(df,beta):
  x = df.iloc[:,:-1]
  y = df.iloc[:,-1]
  for iteration in range(5000):
    print(iteration)
    z = np.dot(x,beta)
    for j in range(beta_count):
      t_dJ=0
      for i in range(n):
        hypo = 1/(1+np.exp(-z[i]))
        dJ = (hypo - y[i])*x.iloc[i,j]
        t_dJ+=dJ
      beta[j]-=(alpha/n)*t_dJ 
  return beta
```

```python
beta1 =findBetas(df1,beta1)
beta2 =findBetas(df2,beta2)
beta3 =findBetas(df3,beta3)
```

```python
print(beta1)
print(beta2)
print(beta3)
```

```python
def findProb(df,beta):
  x = df.iloc[:,:-1]
  z = np.dot(x,beta)
  fx = 1/(1+np.exp(-z))
  for i in range(n):
    if fx[i]<0.5:
      fx[i]=0
  return fx
```

```python
p1=findProb(df1,beta1)
p2=findProb(df2,beta2)
p3=findProb(df3,beta3)
```

```python
p1[3]
```

```python
y_predict = np.zeros(n)
for i in range(n):
  if p1[i]>=p2[i] and p1[i]>=p3[i]:
    y_predict[i] = 0
  elif (p2[i]>=p3[i] and p2[i]>=p1[i]):
    y_predict[i] = 1
  else:
    y_predict[i] = 2
```

```python
def findAccuracy(x,y):
  count = 0
  for i in range(n):
    if(x[i]==y[i]):
      count+=1
  accuracy = (count/n)*100
  print(accuracy)
```

```python
findAccuracy(y,y_predict)
```
