#### Replacer function to replace the string value of numbers by its numeric form

```python
def replacer(data_new_1,column_name,val_to_replace,new_val):
    data_new_1 = data_new_1.replace({
        column_name : val_to_replace
    },new_val)
    
    return data_new_1
```
